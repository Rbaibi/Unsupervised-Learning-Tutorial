Dimension reduction
Dimension reduction removes less-informative "noise" features and  retains the more important and informative features
This makes it easier to analyze and makes computation more efficient.


Principal Component Analysis

First step is decorrelation
PCA Rotates data samples to be aligned with axes
Shifts data samples so they have mean 0
Every row will be a sample

```python
from sklearn import datasets
iris = datasets.load_iris()
samples = iris.data
species = iris.target


#Decorrelating your data
#Visualizing the PCA transformation
from sklearn.decomposition import PCA
model = PCA()
model.fit(samples)
transformed = model.transform(samples)
print(transformed)
print(model.components_)
```

Intrinsic dimension
The intrinsic dimension is the most compact representation of the dataset
First we will calculate the variences of the PCA-features

```python
from sklearn import datasets
iris = datasets.load_iris()
samples = iris.data
species = iris.target

#Plotting the variances of PCA features
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(samples)
features = range(pca.n_components_)

plt.bar(features, pca.explained_variance_)
plt.xticks(features)
plt.ylabel('variance')
plt.xlabel('PCA feature')
plt.show()
```

Reducing dimensions 
Dimension reduction
In this example we are going to keep 2 features

```python
from sklearn import datasets
iris = datasets.load_iris()
samples = iris.data
species = iris.target

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(samples)
transformed = pca.transform(samples)
print(transformed.shape)

import matplotlib.pyplot as plt
xs = transformed[:,0]
ys = transformed[:,1]
plt.scatter(xs, ys, c=species)
plt.show()
```
